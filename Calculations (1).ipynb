```python
import circle
import rectangle_square

def main():
    choice = int(input("Enter 1 to calculate the area and circumference of a circle, 2 to calculate the area of a rectangle, and 3 to calculate the area of a square: "))

    if choice == 1:
        circle.main()
    elif choice == 2:
        rectangle_square.main()
    elif choice == 3:
        rectangle_square.main()
    else:
        print("Invalid choice.")

if __name__ == "__main__":
    main()
```
