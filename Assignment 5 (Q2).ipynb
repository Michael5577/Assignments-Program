```python
def avg_grade(student_name, exam_grades):
  return sum(exam_grades) / len(exam_grades)

def letter_grade(avg_grade):
  if avg_grade >= 90:
    return "A"
  elif avg_grade >= 80:
    return "B"
  elif avg_grade >= 70:
    return "C"
  elif avg_grade >= 60:
    return "D"
  else:
    return "F"

student_names = []
exam_grades = []
for i in range(4):
  student_names.append(input("Enter the name of student {}: ".format(i + 1)))
  exam_grades.append([])
  for j in range(3):
    exam_grades[i].append(float(input("Enter the grade for exam {} of student {}: ".format(j + 1, student_names[i]))))

average_grades = []
for i in range(4):
  average_grades.append(avg_grade(student_names[i], exam_grades[i]))

for i in range(4):
  print("The average grade of {} is {}.".format(student_names[i], average_grades[i]))
  print("The letter grade of {} is {}.".format(student_names[i], letter_grade(average_grades[i])))
```
