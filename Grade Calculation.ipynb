```python
 def calculate_grade(marks):
  """Calculates the grade of a student based on their marks.

  Args:
    marks: A list of marks obtained by the student in different subjects.

  Returns:
    The grade of the student.
  """

  total_marks = sum(marks)
  average = total_marks / len(marks)

  if average >= 90:
    grade = "A"
  elif average >= 80:
    grade = "B"
  elif average >= 70:
    grade = "C"
  elif average >= 60:
    grade = "D"
  else:
    grade = "F"

  return grade


# Example usage:

marks = [85, 90, 75, 80, 95]
grade = calculate_grade(marks)

print("The grade of the student is:", grade)
```
