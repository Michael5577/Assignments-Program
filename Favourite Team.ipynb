```python
# Script to display all the records for the player

# Function to display all the records for the player
def display_player_records(sport, team_name):
    with open(f"{team_name}_{sport}_stats.txt", "r") as file:
        print(file.read())

# Allowing users to display all the records for the player
sport = input("Enter the sport: ")
team_name = input("Enter the team name: ")
display_player_records(sport, team_name)
```
