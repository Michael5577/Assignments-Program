```python
def display_team_statistics(sport, team_name):
  """Displays the statistics of a team from a .txt file.

  Args:
    sport: The sport of the team.
    team_name: The name of the team.
  """

  with open(f"{sport}_{team_name}.txt", "r") as f:
    lines = f.readlines()
    print(lines[0])
    for line in lines[1:]:
      print(line.strip())


# Example usage:

sport = input("What sport is your team? ")
team_name = input("What is the name of your team? ")

# Display the statistics from the .txt file
display_team_statistics(sport, team_name)
```
