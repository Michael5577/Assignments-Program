```python
def store_team_statistics(sport, team_name, statistics):
  """Stores the statistics of a team in a .txt file.

  Args:
    sport: The sport of the team.
    team_name: The name of the team.
    statistics: A list of statistics for the team.
  """

  with open(f"{sport}_{team_name}.txt", "w") as f:
    f.write(f"{sport} {team_name}\n")
    for statistic in statistics:
      f.write(f"{statistic}\n")


# Example usage:

sport = input("What sport is your team? ")
team_name = input("What is the name of your team? ")

# Get the number of statistics the user wants to input
num_statistics = int(input("How many match statistics do you want to store for the player? "))

# Create a list to store the statistics
statistics = []

# Get the statistics from the user
for i in range(num_statistics):
  statistic = input(f"Game {i + 1}: ")
  statistics.append(statistic)

# Store the statistics in a .txt file
store_team_statistics(sport, team_name, statistics)
```
